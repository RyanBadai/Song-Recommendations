# Import Library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from scipy import stats
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.model_selection import GridSearchCV, train_test_split
```

# Load Dataset

```python
df = pd.read_csv("breast-cancer.csv")
df.head()
```

# Clean Dataset

### Check Null Value

```python
df.isna().sum()
```

### Outliers in Dataset

```python
numeric_column = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 
                  'compactness_mean']

plt.figure(figsize = (12, 6))
sns.boxplot(data = df[numeric_column])
plt.title('Box Plot for Numeric Column')
plt.show()
```

```python
numeric_column = ['perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se', 
                  'concave points_se']

plt.figure(figsize = (12, 6))
sns.boxplot(data = df[numeric_column])
plt.title('Box Plot for Numeric Column')
plt.show()
```

```python
numeric_column = [ 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst',
                  'perimeter_worst', 'area_worst']

plt.figure(figsize = (12, 6))
sns.boxplot(data = df[numeric_column])
plt.title('Box Plot for Numeric Column')
plt.show()
```

```python
numeric_column = ['smoothness_worst', 'compactness_worst', 'concavity_worst',  'concave points_worst', 
                  'symmetry_worst', 'fractal_dimension_worst']

plt.figure(figsize = (12, 6))
sns.boxplot(data = df[numeric_column])
plt.title('Box Plot for Numeric Column')
plt.show()
```

```python
outlier_column = ['area_mean', 'radius_se', 'texture_se', 'area_se', 'area_worst', 'compactness_worst',
                 'concavity_worst', 'symmetry_worst']

for column in outlier_column:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    df = df[~((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR)))]

df.shape
```

# Correlation Matrix in Dataset

```python
plt.figure(figsize=(25, 21))
correlation_matrix = df.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, )
plt.title("Correlation Matrix ", size=50)
```

# Fix Value in Dataset

```python
df["diagnosis"] = df["diagnosis"].replace({'M': 0, 'B': 1})
df.head()
```

# Delete ID Column

```python
df = df.drop(columns=['id'])
df.head()
```

# Split The Dataset

```python
x = df.drop(columns = 'diagnosis')
y = df.diagnosis
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
print("Unique classes in y_train:", np.unique(y_train))
```

# Hyperparameter Tunning

```python
parameters = {'C': [0.1, 1, 10],
              'kernel': ['linear', 'rbf'],
              'gamma': [1, 0.1, 0.001]}
```

```python
svm_model = svm.SVC()

grid_search = GridSearchCV(SVC(), parameters, verbose=3)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

# Performance of The Model

### Accuracy

```python
y_pred = grid_search.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred)
print(accuracy_svm)
```

### Confusion Matrix

```python
svm_classification_report = classification_report(y_test, y_pred)
print(svm_classification_report)
```
